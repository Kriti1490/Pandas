```python
# Filtering a DF - Step 1: Create a new dataframe

import pandas as pd

d1 = pd.Series([1001,1002,1003,1004,1005])
d2 = pd.Series(['Mumbai','Bangalore','Delhi','Chennai','Mumbai'])
d3 = pd.Series([10, 23, 15, 17, 11])
d4 = pd.Series([123, 543, 220, 190, 269])
d5 = pd.Series([38, 40, 30, 45, 28])
d6 = pd.Series(['Male','Male','Female','Female','Female'])
columns = {'Cust_ID':d1, 'City':d2, 'Store_no':d3, 'Sales(INR)':d4, 'cust_age':d5, 'Gender':d6}

purchase = pd.DataFrame(columns)
purchase
```

```python
#adding another column

purchase['Salary'] = purchase['Sales(INR)'] - (5/100)*purchase['Sales(INR)']
```

```python
purchase
```

```python
#To display records of only male

purchase[purchase['Gender'] == 'Male']
```

```python
purchase[purchase.Gender == 'Male']
```

```python
#To display records with 2 conditions

purchase[(purchase.Gender == 'Male') & (purchase['Sales(INR)'] < 200)]
```

```python
#another way of applying conditions

cond1 = purchase.Gender == 'Male'
cond2 = purchase['Sales(INR)'] < 200

purchase[cond1 & cond2] #AND condition
```

```python
purchase[cond1 | cond2] #OR condition
```

```python
purchase[~ cond2] #NOT condition
```

```python
cities = ['Bangalore','Mumbai']
purchase[purchase.City.isin(cities)]
```

```python
#Joining

import pandas as pd
e1 = pd.Series([1001,1002,1003,1004,1005])
e2 = pd.Series(['Kriti','Aishwarya','Pranshi','Nikita','Neha'])
e3 = pd.Series(['Noida','Noida','Kanpur','Meerut','Noida'])
e4 = pd.Series([33, 23, 21, 28, 32])

col = {'Emp_id':e1,'Name':e2,'Location':e3,'Age':e4}

employee = pd.DataFrame(col)
employee
```

```python
import pandas as pd
y1 = pd.Series([1001,1002,1003,1004,1008])
y2 = pd.Series([10,20,25,30,35])

col1 = {'Emp_id':y1, 'Years_of_Service':y2}

yos = pd.DataFrame(col1)
yos
```

```python
#Leftjoin

pd.merge(employee,yos, on = 'Emp_id', how = 'left')
```

```python
pd.merge(employee,yos, on = 'Emp_id', how = 'right')
```

```python
emp = pd.merge(employee,yos, on = 'Emp_id', how = 'outer')
emp
```

```python
pd.merge(employee,yos, on = 'Emp_id', how = 'inner')
```
